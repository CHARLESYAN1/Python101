```python
import requests
import time
from scrapy import Selector
import os 
```

```python
def _get_data(url, _type, year, current_page):
    print(url)
    r = requests.get(url)
    selector = Selector(text=r.text).xpath("*//table/tr")
    if selector:
        for col in selector:
            href = col.xpath("th/a/@href")
            title = col.xpath("th/a/@title")
            if href:
                href = href.extract()[0]
                file_format = href.split("/")[-1].split(".")[1].split("?")[0]
                title = title.extract()[0]
                print(href)
                os.system("wget {} --output-document=../data/{}/{}/{}.{}".format(href, _type, year, title, file_format))
                time.sleep(0.5)
    if current_page == 1:
        total_pages = Selector(text=r.text).xpath('*//div[@class="PagesBox"]/table/tbody//a//text()')
        if total_pages:
            total_pages = total_pages.extract()
            print(total_pages)
            max_page = 1
            for p in total_pages:
                try:
                    if int(p)> max_page:
                        max_page = int(p)
                except:
                    pass
            for i in range(2, max_page+1):
                url = "http://irm.cninfo.com.cn/ircs/interaction/irmInformationList.do?irmType={}&beginDate={}-01-01&endDate={}-12-31&pageNo={}".format(_type, year, year, i)
                _get_data(url, _type, year, i)
                
```

```python
_types = ["251310","251311","251312","251313"]

for _type in _types:
    if not os.path.exists("../data/{}/".format(_type)):
        os.mkdir("../data/{}/".format(_type))
    for year in range(2012, 2018):
        if not os.path.exists("../data/{}/{}/".format(_type, year)):
            os.mkdir("../data/{}/{}/".format(_type, year))
        url = "http://irm.cninfo.com.cn/ircs/interaction/irmInformationList.do?irmType={}&beginDate={}-01-01&endDate={}-12-31".format(_type, year, year)
        _get_data(url, _type, year, 1)
```
