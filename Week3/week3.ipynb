### line chart

```python

import matplotlib.pyplot as plt 
from matplotlib import style
style.use("fivethirtyeight")

x = [1,2,3,4,5,6]
y1 = [5,6,3,2,1,7]
y2 = [3,4,1,2,5,3]

plt.plot(x, y1, label="Line1")
plt.plot(x, y2, label="Line2")
plt.title("Interesting Graph\ncheck it out")

plt.xlabel("time")
plt.ylabel("number")
plt.legend()

plt.show()
```

### Bar chart

```python
x = [1,2,3,4,5,6]
y = [5,6,3,2,1,7]

plt.bar(x, y, label="bar chart")
plt.legend()
plt.show()
```

### histogram 

```python
population_ages = [10, 13, 24, 45, 45, 6, 76,2, 3, 44, 45, 86, 19]
bins = [0,20,40,60,80,100]
plt.hist(population_ages, bins, histtype="bar",rwidth=0.2)
plt.title("Hist")
plt.show()
```

### scatter

```python
x = [1,2,3,4,5,6]
y = [5,6,3,2,1,7]
plt.scatter(x, y, color="#d35058", marker="*", s=200) # hex color 
# marker options: https://matplotlib.org/api/markers_api.html

plt.show()
```

### Pie Chart

```python
slices = [7, 3, 4, 12]

activities = ["programming", "gaming","eating", "sleeping"]
cols = ["#ff8f40", "#d35058","#b499cd", "#4d61c1"]

plt.pie(slices,
        labels=activities, 
        colors = cols, 
        startangle=90,
        shadow=True, 
        explode=(0.2, 0.1, 0, 0), 
        autopct="%9.2f%%",
       )
plt.legend(loc=1, prop={'size': 9})
plt.title("Pie chart")
plt.show()
```

### read file and plot

```python
xs = []
ys = []

fin = open("test.txt","r").read()
for l in fin.split("\n"):
    if len(l.split(",")) == 2: 
        x, y = l.split(",")
        xs.append(x)
        ys.append(y)
        
print(xs)
print(ys)

plt.plot(xs, ys, color ="y")
plt.show()
```
